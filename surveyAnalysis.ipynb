```python
#opening csv file
import csv
resultCsv = 'surveyResponses.csv'
rawResult = csv.reader(open('surveyResponses.csv', 'r'), delimiter = ':')

#standard answers for each question
standard = {
    "sProfessions": ["","Artista", "Educador/a", "Gerente de espaco cultural", "Produtor/a cultural", "Tecnica", "Outra"],
    "sWorkArea": ["","Literatura", "Teatro", "Performance", "Circo", "Fotografia", "Danca", "Musica", "Cinema", "Artes visuais", "Novas midias / arte digital"],
    "sSocialMedia": ["","deviantart","flickr","artstation","behance","tiktok","nao faco uso de redes sociais como ferramenta de trabalho","Facebook", "Instagram", "WhatsApp", "Pinterest", "Youtube", "Telegram", "Twitter", "Vimeo", "LinkedIn"],
    "sPortfolio": ["","Redes sociais","Plataforma de video", "Blog pessoal", "Plataformas especializadas em portfolio artistico", "Nuvem", "Site pessoal", "Plataforma de musica", "Meu portfolio nao esta disponivel"],
    "sPortReason": ["","Nunca pensei","Tempo", "Organizacao", "Plataforma", "Privacidade", "Conhecimento", "Necessidade"],
    "sFollow": ["","twitter","youtube","Instagram", "Facebook", "Sites pessoais", "Blogs", "Plataformas de streaming de musica", "Plataformas de streaming de video", "Nao acompanho", "Plataformas de portfolio artistico"],
    "sEvents": ["","Ferramenta de eventos do Facebook", "Sympla", "Eventbrite", "Postagem no Facebook", "Historias do Instagram", "Postagem no Instagram", "Material impresso", "Aplicativos de mensagem", "Imprensa", "Outdoors", "E-mails"],
    "sFunding": ["","Clientes","Universidade","Editais", "Investimento proprio", "Patrocinio", "Ingresso", "Financiamento coletivo"],
    "sScale": ["","Ate 100", "100+", "500+", "1000+", "5000+"],
    "sFollowEvents": ["","sites pessoais","sympla", "Twitter", "Cartazes", "Redes sociais", "WhatsApp", "Facebook", "Indicacao", "Imprensa", "Instagram", "Site de busca", "Blogs", "Sites especializados em eventos", "Email"],
    "sMostImportant": ["","internet","redes sociais", "facebook", "instagram", "whatsapp", "televisao", "radio", "imprensa", "youtube", "indicacao"]
}
sIndex = ["sProfessions","sWorkArea","none","none","sSocialMedia","sSocialMedia","none","sPortfolio","sPortReason","sFollow","sEvents","none","sFunding","sScale","sFollowEvents","sMostImportant","none","none","none", "none"]

for category in standard:
    for i in range(len(standard[category])):
        standard[category][i] = standard[category][i].lower().strip()
# print(standard)

#creating a list containing each person's responses
singleResponse = []

# for each row in the table
for row in rawResult:
    # get the content of each cell
    for cell in row:
        cellCopy=cell
        #check if there are standard values for question
        sValues = sIndex[row.index(cell)]
        
        #while there are any parenthesis, find the index of the opening
        #and closing, then compose the substring in between them and
        #delete it from the cell
        while "(" in cell:
            lPar = cell.find("(")
            rPar = cell.find(")")
            
            snip = cell[lPar:rPar+1]
#           print(snip)
            cell = cell.replace(snip, "")
        

        if sValues != "none":
            cell=cell.split(",")
#             print(cell)
        
            for i in range(len(cell)):
                cell[i]=cell[i].strip().lower()
                if cell[i] not in standard[sValues]:
                    cell[i] = "outra"
                    
            row[row.index(cellCopy)]=",".join(cell)
#         print(cell)
#     print(row)
    #append clean material to the list of answers
    singleResponse.append(row)

print(singleResponse)
```

```python
def analyze (rawResponse):
    # get number of people who responded

    totalPeople =  0
    for line in rawResponse.split("\n"):
        if line is not "":
            totalPeople+=1
    print(totalPeople)
    
    #create list of responses
    responses = rawResponse.replace("\n",",").lower().split(",")
    
    # get number of responses
    totalResponses = len(responses)
    
    # get categories, create an object to contain their results, get absolute number of people who responded each item
    result = {}

    for response in responses:
        response = response.strip()
        if response != "":
            if response in result:
                result[response]["absolute"] += 1
            else:
                result[response] = {"absolute": 1} 

    # percentage of people who responded that (category here corresponds to response in last cell)
    for category in result:
        result[category]["people"] = (result[category]["absolute"]*100)/totalPeople
        
     # percentage of equal responses related to all responses
    for category in result:
        result[category]["response"] = (result[category]["absolute"]*100)/totalResponses

    return(result)
```

```python
#creating a string with all the answers for each question
profession = ""
workField = ""
region = ""
age = ""
socialMedia = ""
workMedia = ""
wouldBeIn = ""
portfolio = ""
portfolioReason = ""
followArt = ""
eventTool = ""
paidAd = ""
funding = ""
scale = ""
audience = ""
    
for r in singleResponse:
    profession += r[0] + "\n"
    workField += r[1] + "\n"
    region += r[2] + "\n"
    age += r[3] + "\n"
    socialMedia += r[4] + "\n"
    workMedia += r[5] + "\n"
    wouldBeIn += r[6] + "\n"
    portfolio += r[7] + "\n"
    portfolioReason += r[8] + "\n"
    followArt += r[9] + "\n"
    eventTool += r[10] + "\n"
    paidAd += r[11] + "\n"
    funding += r[12] + "\n"
    scale += r[13] + "\n"
    audience += r[14] + "\n"
```

```python
# separate states into regions
regionName = ["Sul", "Sudeste", "Centro-oeste", "Norte", "Nordeste"]
statesByRegion = [["PR", "RS", "SC"],["SP", "MG", "ES", "RJ"],["MT", "GO", "MS", "DF"],["AC", "AM", "RO", "RR", "PA", "AP", "TO"],["MA", "PI", "BA", "SE", "AL", "PE", "PB", "RN", "CE"]]

for i in range(len(regionName)):
    for state in statesByRegion[i]:
        region = region.replace(state, regionName[i])
                  
# print(region)
```

```python
surveyResults = {}
surveyResults["profissão"] = analyze(profession)
surveyResults["área de trabalho"] = analyze(workField)
surveyResults["região"] = analyze(region)
surveyResults["idade"] = analyze(age)
surveyResults["redes sociais que utiliza"] = analyze(socialMedia)
surveyResults["redes sociais para trabalho"] = analyze(workMedia)
surveyResults["estaria se não por trabalho"] = analyze(wouldBeIn)
surveyResults["possui portfólio online"] = analyze(portfolio)
surveyResults["se não, motivo"] = analyze(portfolioReason)
surveyResults["acompanha outros artistas"] = analyze(followArt)
surveyResults["divulgação de eventos"] = analyze(eventTool)
surveyResults["paga por anúncios"] = analyze(paidAd)
surveyResults["financiamento"] = analyze(funding)
surveyResults["escala"] = analyze(scale)
surveyResults["eventos como público"] = analyze(audience)

print(surveyResults)
```

```python
# exporting data to a .csv file

# open the file
data = csv.writer(open('rData.csv', 'w', newline = ''), delimiter=':')

# write categories at the top of the file
data.writerow(["Pergunta", "Resposta", "Quantidade", "% de pessoas", "% de respostas"])

# for each question, add a row to identify
for question in surveyResults:
    data.writerow([question.upper()])
    
    # write each category and results
    for category in surveyResults[question]:
        c = surveyResults[question][category]
        data.writerow(["", category, c["absolute"], c["people"], c["response"]])
        
print("Your file has been updated")
```
